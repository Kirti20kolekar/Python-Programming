# Mini Project: Iris Flower Classification (Without ML)

üîç Objective:
Analyze the famous Iris dataset and draw insights like:
Average petal/sepal length
Which species has the longest petals
Simple visualization using Matplotlib/Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Iris dataset from seaborn
df = sns.load_dataset('iris')
```

```python
# Display first 5 rows
print(df.tail())
```

```python
#Explore the Data
print("\nBasic Info:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

print("\nUnique Species:")
print(df['species'].unique())
```

```python
#Group and Analyze
# Average petal length per species
avg_petal = df.groupby('species')['petal_length'].mean()
print("\nAverage Petal Length by Species:")
print(avg_petal)

# Species with longest petals
longest_petal_species = avg_petal.idxmax()
print(f"\n Species with the longest petals: {longest_petal_species}")
```

```python
#Visualize the Data
#Scatter plot
sns.scatterplot(data=df, x='sepal_length', y='petal_length', hue='species')
plt.title("Petal vs Sepal Length by Species")
plt.show()

# Box plot
sns.boxplot(x='species', y='petal_length', data=df, hue='species')
plt.title("Petal Length Distribution")
plt.show()
```
